### **FILL MISIING VALUES**

```python
imputer = KNNImputer(n_neighbors=5)  # temporary default
feature_imputed = imputer.fit_transform(feature_sort)

feature_imputed = pd.DataFrame(feature_imputed, columns=cols)

print("Total missing values after imputation:", np.isnan(feature_imputed.values).sum())
```

```python
# Introduce artificial missingness to test imputation quality
def evaluate_k(feature_df, k, n_splits=3):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    errors = []

    for train_idx, test_idx in kf.split(feature_df):
        train_data = feature_df.iloc[train_idx].copy()
        test_data = feature_df.iloc[test_idx].copy()

        # Store original values at positions to be masked
        mask = np.random.rand(*test_data.shape) < 0.05
        true_values = test_data.values[mask] # Get numpy values

        # Introduce artificial NaNs
        test_data.values[mask] = np.nan # Modify numpy values

        imputer = KNNImputer(n_neighbors=k)
        # Fit on training data and transform test data
        imputed_test_data = imputer.fit(train_data).transform(test_data)

        # Extract imputed values at the masked positions
        imputed_values_at_mask = imputed_test_data[mask]

        # Ensure no NaNs are in the values used for RMSE calculation
        # Find indices where both true_values and imputed_values_at_mask are not NaN
        valid_indices = ~np.isnan(true_values) & ~np.isnan(imputed_values_at_mask)

        if np.any(valid_indices):
            rmse = np.sqrt(mean_squared_error(true_values[valid_indices], imputed_values_at_mask[valid_indices]))
            errors.append(rmse)
        else:
             # Handle the case where no valid comparisons can be made
             errors.append(np.nan) # Append NaN if no valid RMSE can be computed

    # Filter out NaN errors if any occurred
    valid_errors = [e for e in errors if not np.isnan(e)]

    if valid_errors:
        return np.mean(valid_errors)
    else:
        return np.nan # Return NaN if no valid RMSE could be computed across any fold

# Define the objective function for Optuna
def objective(trial):
    k = trial.suggest_int('n_neighbors', 2, 10)  # Suggest integer values for k between 2 and 10
    rmse = evaluate_k(feature_sort, k)
    return rmse

# Ensure feature_sort is a DataFrame and contains original NaNs before this evaluation
# Based on notebook state, feature_sort is currently a standardized DataFrame with NaNs
# So we can proceed directly with feature_sort.

# Create a study object and optimize the objective function
study = optuna.create_study(direction='minimize')  # We want to minimize RMSE
study.optimize(objective, n_trials=20)  # Run 20 trials

# Print the best parameters and the corresponding RMSE
best_k = study.best_params['n_neighbors']
best_rmse = study.best_value

print(f'\n✅ Best k value (from Optuna) = {best_k} (lowest RMSE = {best_rmse:.4f})')
```

```python
best_imputer = KNNImputer(n_neighbors=best_k)
feature_final = best_imputer.fit_transform(feature_sort)
feature_final = pd.DataFrame(feature_final, columns=cols)
print("Final missing values after best imputation:", np.isnan(feature_final.values).sum())
```

```python
# Calculate IQR for Feature_1
Q1_1 = feature_final['Feature_1'].quantile(0.25)
Q3_1 = feature_final['Feature_1'].quantile(0.75)
IQR_1 = Q3_1 - Q1_1

# Define bounds for outliers in Feature_1
lower_bound_1 = Q1_1 - 1.5 * IQR_1
upper_bound_1 = Q3_1 + 1.5 * IQR_1

# Cap outliers in Feature_1
feature_final['Feature_1'] = np.where(
    feature_final['Feature_1'] < lower_bound_1,
    lower_bound_1,
    np.where(
        feature_final['Feature_1'] > upper_bound_1,
        upper_bound_1,
        feature_final['Feature_1']
    )
)

# Calculate IQR for Feature_9
Q1_9 = feature_final['Feature_9'].quantile(0.25)
Q3_9 = feature_final['Feature_9'].quantile(0.75)
IQR_9 = Q3_9 - Q1_9

# Define bounds for outliers in Feature_9
lower_bound_9 = Q1_9 - 1.5 * IQR_9
upper_bound_9 = Q3_9 + 1.5 * IQR_9

# Cap outliers in Feature_9
feature_final['Feature_9'] = np.where(
    feature_final['Feature_9'] < lower_bound_9,
    lower_bound_9,
    np.where(
        feature_final['Feature_9'] > upper_bound_9,
        upper_bound_9,
        feature_final['Feature_9']
    )
)

print("Outliers in 'Feature_1' and 'Feature_9' have been capped using the IQR method.")
```

### **ADJACENCY MATRIX CREATION**

```python
# --- Step 2: Count connections per source node ---
connection_count = edge.groupby("id_1")["id_2"].nunique().reset_index()

# Rename columns for clarity
connection_count.columns = ["source_id", "num_connections"]

# --- Step 3: Sort (optional, by most connected nodes) ---
connection_count = connection_count.sort_values(by="num_connections", ascending=False)

# --- Step 4: Display and/or save ---
print("✅ Source-to-target connection counts:")
print(connection_count.head(10))  # show top 10 most connected
```

```python
# --- Step 2: Extract unique node IDs ---
nodes = pd.unique(edge[['id_1', 'id_2']].values.ravel())
nodes = sorted(nodes)
node_to_idx = {node: idx for idx, node in enumerate(nodes)}

num_nodes = len(nodes)
print(f"Total unique nodes: {num_nodes}")

# --- Step 3: Build adjacency indices ---
row_indices = [node_to_idx[src] for src in edge['id_1']]
col_indices = [node_to_idx[dst] for dst in edge['id_2']]

# Undirected graph → add reverse edges
row_indices += col_indices
col_indices += row_indices[:len(edge)]

# --- Step 4: Create sparse adjacency matrix ---
indices = torch.tensor([row_indices, col_indices], dtype=torch.long)
values = torch.ones(indices.shape[1])
adjacency = torch.sparse_coo_tensor(indices, values, (num_nodes, num_nodes))

# --- Step 5: Convert to dense matrix (⚠️ only for small graphs) ---
adj_dense = adjacency.to_dense()
adj_df = pd.DataFrame(adj_dense.numpy(), index=nodes, columns=nodes)

# --- Step 6: Save compressed adjacency matrix ---
output_path = "/content/facebook_adjacency_matrix.csv.gz"
adj_df.to_csv(output_path, index=True, compression="gzip")

print(f"✅ Compressed adjacency matrix saved as '{output_path}'")
print("Shape:", adj_df.shape)
```
